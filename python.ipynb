```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn

df=pd.read_csv("customer.csv")
df
```

```python
df.info()
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
    if value==1:
        return "yes"
    else:
        return "no"
    
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
a = sbn.countplot(x="Churn", data=df)
a.bar_label(a.containers[0])
plt.title("count of customers by churn")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"], labels =gb.index, autopct ="%1.2f%%")
plt.show()
```

```python
plt.figure(figsize=(3,3))
sbn.countplot(x="gender" , data = df ,hue="Churn")
plt.show()
```

```python
plt.figure(figsize=(3,3))
a = sbn.countplot(x="SeniorCitizen",data=df)
a.bar_label(a.containers[0])
plt.title("count of customer by senior citizen")
plt.show()
```

```python
plt.figure(figsize=(9,3))
sbn.histplot(x="tenure" , data=df , bins=50)
plt.show()
```

```python
plt.figure(figsize=(4,4))
ax = sbn.countplot(x="Contract",data=df , hue="Churn")
ax.bar_label(ax.containers[0])
plt.title("count of customers by contract")
plt.show()
```

```python
# Example: load your DataFrame
# df = pd.read_csv('your_data.csv')

# List of columns to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create subplots
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 12))

# Flatten axes array for easy iteration
axes = axes.flatten()

for i, col in enumerate(columns):
    sbn.countplot(data=df, x=col, ax=axes[i], hue =df["Churn"] )
    axes[i].set_title(f'Countplot of {col}')
    axes[i].tick_params(axis='x', rotation=45)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
